```python
import numpy as np
import pandas as pd
import pandas_profiling as pdp
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
import json
sns.set_style('darkgrid')
%matplotlib inline
from tqdm import tqdm
```

```python
train = pd.read_csv('train.csv', index_col='row_id')
# train.drop_duplicates(keep='first', inplace=True)
train.reset_index(drop=True, inplace=True)
test = pd.read_csv('test.csv', index_col='row_id')
# data = pd.concat([train, test], sort=False)
# train = data[:len(train)]
# test = data[len(train):]
```

```python
data = pd.concat([train, test])
train_duplicated = train[data.duplicated(subset=test.columns)[:len(train)]]
test_duplicated = test[data.duplicated(subset=test.columns)[len(train):]]
```

```python
train_duplicated
```

```python
test_duplicated
```

```python
df = pd.DataFrame()
df['index'] = test_duplicated.index
df['index'] = df['index'].astype('int')
df['target'] = np.nan
df
```

```python
with tqdm(range(len(df))) as pbar:
    for i, row in df.iterrows():
        statement = "".join([" {}=={} & ".format(c, test.loc[int(row['index']), c]) for c in test.columns])
        statement += "1==1"
        q = train_duplicated.query(statement)
        if not q.empty:
            df.iloc[i, 1] = q['target'].iloc[-1]
        pbar.update(1)
```
